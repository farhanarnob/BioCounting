```python
def hammingDistance(string1, string2): 
    distance = 0
    L = len(string2)
    for i in range(L):
        if string1[i] != string2[i]:
            distance += 1
    return distance
        

def neighbours(pattern,distance):
    nucleotides = ["A","C","G","T"]
    if (distance == 0) : 
        return list(pattern)
    

    if (len(pattern)==0) :
        return list()
    

    if (len(pattern)==1) :
        return nucleotides

    neighbourhood = list()
    suffixOfPattern = pattern[1:]
    suffixNeighbor = neighbours(suffixOfPattern,distance)
    for s in suffixNeighbor:
        distanceValue = hammingDistance(suffixOfPattern,s)
        if(distanceValue < distance):
            for n in nucleotides:
               neighbourhood.append(f"{n}{s}")
        else:
            neighbourhood.append(f"{pattern[0]}{s}")
    return neighbourhood
print("\n".join(neighbours("TCATAAGCCA",1)))
```
