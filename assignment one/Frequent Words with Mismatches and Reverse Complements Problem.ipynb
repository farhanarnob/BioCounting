```python
from collections import defaultdict

def hammingDistance(seqOne,seqTwo):
    distance=0
    for start in range(len(seqOne)): 
      if seqOne[start]!=seqTwo[start] : distance+=1;
    return distance

def reversePattern(pattern):
    temp = pattern.replace("A","X").replace("T","A").replace("X","T")
    return (temp.replace("G","X").replace("C","G").replace("X","C")[::-1]) 
    

def neighbour(pattern,mismatch, words):
    if mismatch == 0 : words.add(pattern)
    else:
        bases = ['A', 'T', 'C', 'G']
        for i in range(len(pattern)):
            for j in range(len(bases)):
                newPattern = pattern[:i] + bases[j] + pattern[i+1:]
                if mismatch <= 1: words.add(newPattern)
                else: neighbour(newPattern, mismatch-1, words)

def findMostFrequentPattern(text, k, d):
    allFrequentwords = defaultdict(int)
    for i in range(len(text) - k + 1):
        frequentwords = set()
        neighbour(text[i:i + k], d, frequentwords)

        for words in frequentwords:
            allFrequentwords[words] += 1

    for t in allFrequentwords.keys():
        reverse_k = reversePattern(t)
        for i in range(len(text) - k + 1):
            if hammingDistance(text[i:i + k], reverse_k) <= d:
                allFrequentwords[t] += 1

    result = set()
    for t in allFrequentwords.keys():
        if allFrequentwords[t] == max(allFrequentwords.values()):
            result.add(t)
            result.add(reversePattern(t))
    for i in result:
        print(i, end=" ")
        
text = "AGTAGCTTCTCACTCCCTAGAGGTAATGCGCAAGGGAGTAGCTTCTGCGCAAGGGCACTCCCTCACTCCCTGGCGTTGTTTAGTAGCTTCTAGAGGTAATGCGCAAGGGAGTAGCTTCTGGCGTTGTTTCACTCCCTAGAGGTAATAGAGGTAATAGTAGCTTCTCACTCCCTAGTAGCTTCTAGAGGTAATCACTCCCTAGAGGTAATGGCGTTGTTTGGCGTTGTTTGCGCAAGGGGCGCAAGGGAGTAGCTTCTAGTAGCTTCTCACTCCCTGCGCAAGGGAGAGGTAATGGCGTTGTTTAGTAGCTTCTGCGCAAGGGGGCGTTGTTTCACTCCCTAGTAGCTTCTAGTAGCTTCTCACTCCCTAGAGGTAATAGTAGCTTCTGGCGTTGTTTGGCGTTGTTTAGTAGCTTCTAGTAGCTTCTGCGCAAGGGAGTAGCTTCTGCGCAAGGGGGCGTTGTTTAGTAGCTTCTAGTAGCTTCTAGTAGCTTCTAGTAGCTTCTAGTAGCTTCTAGAGGTAATGCGCAAGGGGCGCAAGGGGCGCAAGGGAGTAGCTTCTGGCGTTGTTTGCGCAAGGGGCGCAAGGGAGTAGCTTCTGCGCAAGGGGGCGTTGTTTGCGCAAGGGGGCGTTGTTTAGAGGTAATAGAGGTAATGGCGTTGTTTAGAGGTAATCACTCCCTAGAGGTAATGCGCAAGGGGCGCAAGGGCACTCCCTCACTCCCTAGAGGTAATAGTAGCTTCTCACTCCCTAGTAGCTTCTAGTAGCTTCTCACTCCCTAGTAGCTTCTAGAGGTAATAGAGGTAATAGTAGCTTCTAGAGGTAATAGAGGTAATAGTAGCTTCTCACTCCCTGGCGTTGTTTAGAGGTAATAGAGGTAATAGAGGTAATGCGCAAGGGAGTAGCTTCTGCGCAAGGGGCGCAAGGGAGAGGTAATCACTCCCTGGCGTTGTTTCACTCCCTCACTCCCTCACTCCCTGGCGTTGTTT"
k, d = 7,2
findMostFrequentPattern(text, k, d)
```
