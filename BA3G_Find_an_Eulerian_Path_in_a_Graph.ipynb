```python
from random import choice

# convert to adj list
def parse_adj_list(adj_list_text):
    adj_list = {}
    for line in adj_list_text:
        fst_split = line.split(' -> ')
        adj_list[fst_split[0]] = fst_split[1].split(',')
    return adj_list

def remove_edge(adj_list, start_node, target_node):
    # if found, we will first remove the edge from start to target
    adj_list[start_node].remove(target_node)
    if not adj_list[start_node]: # if there is not any associated value with start_node 
        del adj_list[start_node] # delete it from list
    return adj_list

def find_euerian_cycle(adj_list):

    # step 1: form a cycle Cycle by randomly walking in Graph (don't visit the same edge twice!)
    # randomly picking node and corresponding edges
    start_node, edges = choice(list(adj_list.items()))

    # randomly picking one of the target nodes from edgelist
    target_node = choice(edges)

    # as we have visited one edge, we can remove it from our adjList
    adj_list = remove_edge(adj_list, start_node, target_node) 

    # our Cycle
    cycle = [start_node, target_node]   

    # making target node the current node from where we will continue the traversal
    current_node = target_node


    while current_node != start_node:
        # find current node edges
        edges = adj_list[current_node]

        # picking a target node randomly
        target_node = choice(edges)

        # now again we have visited currentNode -> targetNode, so we can remove this edge

        adj_list = remove_edge(adj_list, current_node, target_node)

        # updating currentNode
        current_node = target_node

        # appending new node into cycle
        cycle.append(current_node)

    # now we need to check for any unexpected edges
    # if any, then do the same process starting from unexpected edges
    # otherwise return the cycle
    while adj_list:
         # checking if any node in cycle exist in adjList, storing them into potentialStarts

        potential_starts = [(idx, node) for idx, node in enumerate(cycle) if node in adj_list]

        # choosing randomly a start_point
        idx, new_start = choice(potential_starts)

        # forming new cycle starting from new start and then randomly walking
        new_cycle = cycle[idx:] + cycle[1:idx+1]

        edges = adj_list[new_start]

        # selecting new target node
        target_node = choice(edges)

        adjList = remove_edge(adj_list, new_start, target_node)

        current_node = target_node
        new_cycle.append(current_node)
            # starting randomly traversal until currentNode != newStart
        while current_node != new_start:
            edges = adjList[current_node]
            target_node = choice(edges)
            adjList = remove_edge(adjList, current_node, target_node)
            current_node = target_node
            new_cycle.append(current_node)
            cycle = new_cycle
    return cycle
def find_euerian_path(adj_list):
    deg_diff = {}
    for node, edges in adj_list.items():
        # print(edges)
        # print(len(edges))
        # storing indegree
        if node in deg_diff:
            deg_diff[node] += len(edges)
        else:
            deg_diff[node] = len(edges)
    # print(deg_diff["10"])

    # To findout the outdegree of corresponding edges, if they exist in deg_diff, that means
    # they have indegree assigned by previous if else. otherwise they have no indegree.
    # as #indegree is 1 in all the cases, so we can subtract 1 or assign -1
        for edge in edges:
            if edge in deg_diff:
                deg_diff[edge] -= 1
            else:
                deg_diff[edge] = -1
    for key in deg_diff:
        if(deg_diff[key]==-1):
            ending_node = deg_diff[key]
            print(key+":"+str(ending_node))# this is the end node
        elif(deg_diff[key]==1):
            start_node = deg_diff[key]
            print(key+":"+str(start_node))# this is the start node
    # print(deg_diff)

    # Euler path will exist iff euler cycle is exist
    # if ending node present draw edge from ending to starting node
    if ending_node in adj_list:
        adj_list[ending_node].append(start_node)
  # else create a new edge from ending node to starting node
    else:
        adj_list[start_node] = start_node

    # print(adj_list)
    euler_cycle = find_euerian_cycle(adj_list)
    return adj_list
# main
with open ('/content/eulerian_path (1).txt', 'r') as f:
    input_lines = [line.strip() for line in f.readlines()]
    # print(*input_lines)
    adj_list = parse_adj_list(input_lines)
    # print(*adj_list)
    euerian_path = find_euerian_path(adj_list) 
```
